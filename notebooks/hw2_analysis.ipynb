```python
import matplotlib as mpl
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

plt.style.use('seaborn-paper')
sns.set_style('whitegrid')
```

## Experience 1: CartPole-v0

```python
sb_df = pd.read_excel(
    'data/hw2/exp_01/all_data.xlsx',sheet_name='Small Batch'
)
display(sb_df)
lb_df = pd.read_excel(
    'data/hw2/exp_01/all_data.xlsx', sheet_name='Large Batch'
)
display(lb_df)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4), dpi=224, sharey=True)

axes[0] = sns.lineplot(data=sb_df, x='Step', y='Evaluation Average Return',
                       hue='Name', palette='Set2', ax=axes[0], alpha=0.75)
axes[0].set_title('(a) Small Batch Experiements', fontsize=12)

axes[1] = sns.lineplot(data=lb_df, x='Step', y='Evaluation Average Return',
                       hue='Name', palette='Set2', ax=axes[1], alpha=0.75)
axes[1].set_title('(b) Large Batch Experiments', fontsize=12)
fig.tight_layout()

fig.savefig('../docs/hw2_report/img/exp_01.png', dpi=600)
```

## Experiment 02

```python
df = pd.read_csv(
    'data/hw2/exp_02/run-q2_pg_q2_b500_r0.01_InvertedPendulum-v4_26-09-2022_20-35-21-tag-Eval_AverageReturn.csv',
    sep=',', encoding='utf-8')
df['Step'] = df['Step'] + 1
df.rename(columns={'Value': 'Evaluation Average Return'}, inplace=True)
fig, ax = plt.subplots(1, 1, figsize=(6, 4), dpi=224)
ax = sns.lineplot(data=df, x='Step', y='Evaluation Average Return',
                  alpha=0.75, palette='Set2', ax=ax)
ax.set_title(r'$b^*=500$, $r*=0.01$', fontsize=12)
fig.tight_layout()
fig.savefig('../docs/hw2_report/img/exp_02.png', dpi=600)
```
